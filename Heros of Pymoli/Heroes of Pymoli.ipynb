## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#count number of purchase for top items
itemcount=purchase_data.groupby(['Item ID','Item Name'])["Item Name"].count().sort_values(ascending=False)
popular_item=pd.DataFrame(data=itemcount)

#add information to the dataframe 
popular_item_revise=popular_item.rename(columns={"Item Name":"Purchase Count"})
popular_item_revise["Total Purchase Value"]=purchase_data.groupby(['Item ID','Item Name'])["Price"].sum()
popular_item_revise["Total Purchase Value"]=popular_item_revise["Total Purchase Value"].map('${:,.2f}'.format)
popular_item_revise["Item Price"]=purchase_data.groupby(['Item ID','Item Name'])["Price"].mean()
popular_item_revise["Item Price"]=popular_item_revise["Item Price"].map('${:,.2f}'.format)
popular_item_final=popular_item_revise[["Purchase Count","Item Price","Total Purchase Value"]]
popular_item_final.head()

```
